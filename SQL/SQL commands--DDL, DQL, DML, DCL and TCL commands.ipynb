# SQL commands are mainly divided into five categories: 
   - DDL – Data Definition Language.DDL is a set of SQL commands for creating, modifying, and deleting database structures, but not data. These commands are not typically used by a general user who should access the database through an application.
   
   - DQL - Datenabfragesprache
   
   - DML – Data Manipulation Language
   
   - DCL – Datenkontrollsprache
   
   - TCL – Transaction Control Language
